```python
import os
import csv

assigned_label = 0
train_root = '/Users/ayushmuralidharan/Desktop/AOML/AOML PROJECT/MAML-Pytorch-master/Experiment 2/Dataset_INDIAN_FULL/train'
test_root = '/Users/ayushmuralidharan/Desktop/AOML/AOML PROJECT/MAML-Pytorch-master/Experiment 2/Dataset_INDIAN_FULL/test'
train_alphabets = [d for d in os.listdir(train_root) if os.path.isdir(os.path.join(train_root, d))]
test_alphabets = [d for d in os.listdir(test_root) if os.path.isdir(os.path.join(test_root, d))]

with open('train_data.csv', 'w') as csvfile:
    fields = ['ID', 'Path', 'Label']
    csvwriter = csv.writer(csvfile, delimiter=',')
    csvwriter.writerow(fields)
    for a in train_alphabets:
        for charfolder in os.listdir(os.path.join(train_root, a)):
            charfolder_path = os.path.join(train_root, a, charfolder)
            if os.path.isdir(charfolder_path):  # Only proceed if it is a directory
                for image in os.listdir(charfolder_path):
                    image_path = os.path.join(charfolder_path, image)
                    if os.path.isfile(image_path):  # Ensure it's a file
                        csvwriter.writerow([image, image_path, str(assigned_label)])
        assigned_label += 1

with open('test_data.csv', 'w') as csvfile:
    fields = ['ID', 'Path', 'Label']
    csvwriter = csv.writer(csvfile, delimiter=',')
    csvwriter.writerow(fields)
    for a in test_alphabets:
        for charfolder in os.listdir(os.path.join(test_root, a)):
            charfolder_path = os.path.join(test_root, a, charfolder)
            if os.path.isdir(charfolder_path):  # Only proceed if it is a directory
                for image in os.listdir(charfolder_path):
                    image_path = os.path.join(charfolder_path, image)
                    if os.path.isfile(image_path):  # Ensure it's a file
                        csvwriter.writerow([image, image_path, str(assigned_label)])
        assigned_label += 1
```
